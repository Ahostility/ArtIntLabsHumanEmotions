```python
import cv2
import syspath
import numpy as np
from src.modules.video.face import FaceCapture
from src.modules.video.face.plot import FacemarksPlot, EmotionPlot

facecap = FaceCapture()
```

```python
fplot = FacemarksPlot(facecap)
eplot = EmotionPlot(facecap)
cap = cv2.VideoCapture(0)

try:
    while True:
        ret, frame = cap.read()
        if not ret or cv2.waitKey(1) == 27: break

        facecap.detect(frame, stabilize=True)
        eplot.show_bar(frame)
        fplot.show3d()
        fplot.show2d(frame, landmarks=True, stabilizer=False, gaze=False)

except KeyboardInterrupt: pass

cap.release()
cv2.destroyAllWindows()
```
